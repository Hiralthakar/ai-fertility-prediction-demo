# AI for Fertility: Mini Project
This notebook performs basic analysis on menstrual cycle data.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For prediction
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```
df = pd.read_csv('fertility_dummy_data.csv')
df.head()
```

```
df.describe()
```

```
sns.scatterplot(data=df, x='Cycle_Length_Days', y='Ovulation_Day', hue='User_ID')
plt.title('Cycle Length vs. Ovulation Day')
plt.show()
```

```
# Features and target
X = df[['Cycle_Length_Days', 'Mood_Score', 'Cramping_Intensity']]
y = df['Ovulation_Day']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print(f'Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%')
```
